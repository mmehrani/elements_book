```python
import networkx as nx
import matplotlib.pyplot as plt
import pickle
```

```python
list_of_books_dict = []
for i in range(1,14):
    a_file = open("books_references_dictionaries/ref_in_book_{}.pkl".format(i), "rb")
    dic = pickle.load(a_file)
    list_of_books_dict.append(dic)

list_of_books_dict[0]
```

```python
G = nx.DiGraph(list_of_books_dict[0])
for i in range(len(list_of_books_dict)):
    H = nx.DiGraph(list_of_books_dict[i])
    attrs = {node:{'book':i+1} for node in list_of_books_dict[i].keys()}
    nx.set_node_attributes(H, attrs)
    G = nx.compose(G,H)
```

```python
nx.write_gexf(G,'total_books_graph.gexf')
```
